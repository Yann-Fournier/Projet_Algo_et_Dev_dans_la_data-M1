```python
import os
from GetData import GetData
from sklearn.model_selection import train_test_split
from Utils import build_lstm_model, add_indicators, preprocess
```

```python
# Dictionnaire des timeframes
timeframes = {
    "1h": ["CT"], 
    "2h": ["CT"], 
    "4h": ["CT"], 
    "12h": ["MT"], 
    "1d": ["MT", "LT"], 
    "1w": ["MT", "LT"], 
    "1M": ["LT"], 
}
```

```python
features = [
    'open',
    'high',
    'low',
    'close',
    'volume',
    'rsi',
    'rsima',
    'macd',
    'macd_diff',
    'macd_signal',
    'stoch_%K',
    'stoch_%D',
    'lower_band',
    'higher_band',
    'ma_band',
    'sma9',
    'sma21',
    'Avg_day',
    'Avg_corps',
    'Avg_meches'
]

results = {}

# CrГ©er un dossier pour stocker les modГЁles si non existant
os.makedirs("modeles", exist_ok=True)
```

```python
# Boucle sur les timeframes et leurs labels associГ©s pour entraГ®ner un modГЁle par timeframe
for tf, tab_Label in timeframes.items():
    for label in tab_Label:
        print(f"\nCrГ©ation du modГЁle pour le timeframe : {tf} ({label})")

        # TГ©lГ©chargement et prГ©paration des donnГ©es
        DATA = BTC_USDT_1D = GetData(
            exchange="binance",
            pair="BTC/USDT",
            tf=tf,
        )
        await BTC_USDT_1D.download_data()
        BTC_USDT_1D.data = add_indicators(BTC_USDT_1D.data, tf)

        # PrГ©traitement et crГ©ation des sГ©quences
        X, y = preprocess(DATA.data, features, label="Label_" + label, time_steps=60)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

        # CrГ©ation et entraГ®nement du modГЁle
        print("EntraГ®nement du modГЁle...")
        model = build_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))
        history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

        # Г‰valuation du modГЁle
        loss, acc = model.evaluate(X_test, y_test)
        print()
        print(f"{tf} - Test Accuracy: {acc:.4f}")
        print(f"{tf} - Test Loss: {loss:.4f}")

        # Stockage des rГ©sultats
        results[tf + "_" + label] = {
            'model': model,
            'history': history,
            'accuracy': acc,
            'loss': loss
        }

        # Sauvegarde du modГЁle pour ce timeframe
        model_path = f"modeles/lstm_btc_usdt_{tf}_{label}.h5"
        model.save(model_path)
        print(f"ModГЁle sauvegardГ© : {model_path}")
        print("-" * 100)

# Temps (pour 9 modГЁles): 6m
```
